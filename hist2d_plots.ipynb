```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load image
imagename = 'picture.jfif'
im = imageio.imread(imagename,as_gray=True)
print(im.shape)

coords = []

x = 0
y = 0
for row in im:
    y = 0
    for col in row:
        # print(int(col))
        count = int(col) + 10 # Add 10 to ensure no white space
        for i in range(int(count/10)):
            coords.append([x + np.random.rand() - 0.5, y + np.random.rand() - 0.5])
        y += 1
    x += 1

df = pd.DataFrame(coords)
df.to_csv('data.csv',index=False)
len(df)
```

```python
x = df[1]
y = -df[0]
plt.plot(x,y)
```
